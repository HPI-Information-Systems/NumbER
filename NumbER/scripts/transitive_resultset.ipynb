```python
import pandas as pd
import numpy as np
import os

from NumbER.matching_solutions.utils.transitive_closure import calculate_clusters, calculate_transitive_closure
```

```python
path = "/hpi/fs00/share/fg-naumann/lukas.laskowski/experiments/"
folders = os.listdir(path)
results = []
for result_folder in list(filter(lambda x: "restart" in x, folders)):
    try:
        for experiment in os.listdir(os.path.join(path, result_folder, "matching_solution", "embitto")):
            iteration = experiment.split("_")[-1][:-4]
            dataset = os.listdir(os.path.join(path, result_folder, "datasets"))[0]
            result = pd.read_csv(os.path.join(path, result_folder, "matching_solution", "embitto", f"{dataset}_{iteration}.csv"))
            gt = pd.read_csv(f"/hpi/fs00/share/fg-naumann/lukas.laskowski/datasets/{dataset}/samples/similarity/test_{iteration}_goldstandard.csv")
            gt_clusters = calculate_clusters(gt)
            clusters = calculate_clusters(result)
            print(f"========={dataset}_{iteration}=========")
            cluster_avg = np.mean([len(cluster) for cluster in clusters])
            gt_cluster_avg = np.mean([len(cluster) for cluster in gt_clusters])
            print("Average length of clusters: ", cluster_avg)
            print("Average length of gt clusters: ", np.mean([len(cluster) for cluster in gt_clusters]))
            results.append(cluster_avg-gt_cluster_avg)
            # print(gt)
    except Exception as e:
        print(e)
print(np.mean(results))
```
